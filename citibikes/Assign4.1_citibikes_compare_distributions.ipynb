# BY GENDER

```python
#df is the dataframe where the content of the csv file is stored
df['ageM'] = 2015-df['birth year'][(df['usertype'] == 'Subscriber') & (df['gender'] == 1)]
df['ageF'] = 2015-df['birth year'][(df['usertype'] == 'Subscriber') & (df['gender'] == 2)]
```

# KS TEST

```python
#cumulative distribution
bins = np.arange(10, 99, 10)
df.ageM.groupby(pd.cut(df.ageM, bins)).agg([count_nonzero]).plot(kind='bar')
df.ageF.groupby(pd.cut(df.ageF, bins)).agg([count_nonzero]).plot(kind='bar')
```

We can see that there are more male riders than female riders 200,000 vs. 60,000. The most male riders range from ages 30 - 40, whereas the most female riders are younger from 20 - 30.

```python
#print df.ageS, df.ageS.cumsum()
csM=df.ageM.groupby(pd.cut(df.ageM, bins)).agg([count_nonzero]).cumsum()
csF=df.ageF.groupby(pd.cut(df.ageF, bins)).agg([count_nonzero]).cumsum()

print np.abs(csM / csM.max()-csF / csF.max())

pl.plot(bins[:-1] + 5, csM / csM.max(), label = "M")
pl.plot(bins[:-1] + 5, csF / csF.max(), label = "F")
pl.legend()
```

```python
import scipy.stats
```

# KS tests to compare 2 samples
http://docs.scipy.org/doc/scipy-0.15.1/reference/generated/scipy.stats.ks_2samp.html

```python
ks=scipy.stats.ks_2samp(df.ageM, df.ageF)
print ks
```

# KS test - extra credit

remember: the Null hypothesis is rejected if

$D_KS(n1,n2) > c(\alpha) \sqrt{\frac{(n1+n2)}{n1n2}}$

where $c(\alpha)$ is the inverse of the KS distribution, and you do not have to know how to get that cause there are tables that list critical values!!

http://www.real-statistics.com/tests-normality-and-symmetry/statistical-tests-normality-symmetry/kolmogorov-smirnov-test/kolmogorov-distribution/

```python
n1 = df.ageM[~np.isnan(df.ageM)].count()
n2 = df.ageF[~np.isnan(df.ageF)].count()
print n1, n2
```

```python
ksinv = 1.36*sqrt((n1+n2)*1.0/(n1*n2))

#Null hypothesis is in fact that they are from the same distribution
#using significance level 0.05, c(a) = 1.36
if ks[0] > ksinv:
    print ks[0], ">", ksinv, 'We reject the null hypothesis that the two samples are from the same distribution.'
else:
    print 'We cannot reject the null hypothesis.'
```

# Pearson's test for correlation

http://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.pearsonr.html#scipy.stats.pearsonr

```python
#scipy.stats.pearsonr(df.ageM, df.ageF)
#returns nan

ageM1 = np.array(df.ageM[~np.isnan(df.ageM)])
ageF1 = np.array(df.ageF[~np.isnan(df.ageF)])

aM = np.random.choice(ageM1, size = 5000)
aF = np.random.choice(ageF1, size = 5000)

scipy.stats.pearsonr(aM, aF)
```

This implies no linear relationship correlation because it is between -1 and +1 with 0, therefore we accept the null hypothesis.

# Spearman's test for correlation

http://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.spearmanr.html#scipy.stats.spearmanr

```python
scipy.stats.spearmanr(aM, aF, axis=0)
```

Since it is between -1 and +1 with 0, implies that there's no monotonic correlation relationship.

# THE DISTRIBUTION OF RIDERS DURING THE DAY VS DURING THE NIGHT

hint: remember how I converted the starttime last time in my own citibikes notebook into units that are easy to read. df['mystarttime'] = ... then you can use something like df['mystarttime'].dt.hour

```python
import datetime
df['hours'] = pd.to_datetime(df['starttime']).dt.hour
```

```python
df['dayR'] = 2015-df['birth year'][(df['hours'] > 6) & (df['hours'] <= 18)]
df['nightR'] = 2015-df['birth year'][(df['hours'] > 18) | (df['hours'] <= 6)]
```

```python
bins = np.arange(10, 99, 10)
df.dayR.groupby(pd.cut(df.dayR, bins)).agg([count_nonzero]).plot(kind='bar')
df.nightR.groupby(pd.cut(df.nightR, bins)).agg([count_nonzero]).plot(kind='bar')
```

```python
csM=df.ageM.groupby(pd.cut(df.ageM, bins)).agg([count_nonzero]).cumsum()
csF=df.ageF.groupby(pd.cut(df.ageF, bins)).agg([count_nonzero]).cumsum()
csD=df.dayR.groupby(pd.cut(df.dayR, bins)).agg([count_nonzero]).cumsum()
csN=df.nightR.groupby(pd.cut(df.nightR, bins)).agg([count_nonzero]).cumsum()

print np.abs(csM / csM.max()-csF / csF.max())

pl.plot(bins[:-1] + 5, csM / csM.max(), label = "M")
pl.plot(bins[:-1] + 5, csF / csF.max(), label = "F")
pl.plot(bins[:-1] + 5, csM / csD.max(), label = "D")
pl.plot(bins[:-1] + 5, csF / csN.max(), label = "N")
pl.legend()
```

# KS tests to compare 2 samples

```python
ks2=scipy.stats.ks_2samp(df.dayR, df.nightR)
print ks2
```

```python
n3 = df.ageM[~np.isnan(df.dayR)].count()
n4 = df.ageF[~np.isnan(df.nightR)].count()
print n3, n4

ksinv2 = 1.36*sqrt((n3+n4)*1.0/(n3*n4))

#Null hypothesis is in fact that they are from the same distribution
#using significance level 0.05, c(a) = 1.36
if ks2[0] > ksinv2:
    print ks2[0], ">", ksinv2, 'We reject the null hypothesis that the two samples are drawn from the same distribution.'
else:
    print 'We cannot reject the null hypothesis.'
```
